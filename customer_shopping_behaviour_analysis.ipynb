```python
#instead of using mean or median to fill null values we are using category wise median to fill null values in order to avoid clothing category getting review of footwear category and same goes for other categories as well.
#we are using median instead of mean because median is more robust to outliers.

df['Review Rating']=df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age_group
#qcut splits the data into 4 groups based on data distribution
labels=['Young Adult', 'Adult', 'Middle-Aged', 'Senior']
df['age_group']=pd.qcut(df['age'], q=4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping= {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quaterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```
